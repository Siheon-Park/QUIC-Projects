```python
from sklearn import svm
from datageneration import iris_data
from swap_classifier import Classical_SWAP_Duel_Classifier

import numpy as np 
import matplotlib.pyplot as plt
```

```python
X, y = iris_data()

C=1
clf = svm.SVC(kernel='poly', degree=2, gamma=1, coef0=0, C=C)
clf.fit(X, y)
alpha = np.zeros(len(y))
alpha[clf.support_]=np.abs(clf.dual_coef_)
plt.plot(alpha)
clf.score(X, y)
np.sum(alpha*y)
plt.figure()
swp = Classical_SWAP_Duel_Classifier(X, y)
swp.alpha = np.abs(alpha)
swp.plot()
```

```python
from scipy.optimize import minimize, rosen, rosen_der
x0 = [1.3, 0.7, 0.8, 1.9, 1.2]
res = minimize(rosen, x0, method='Nelder-Mead', tol=1e-6)
res.x
```

```python
res = minimize(rosen, x0, method='BFGS', jac=rosen_der,
               options={'gtol': 1e-6, 'disp': True})
```

```python
fun = lambda x: (x[0] - 1)**2 + (x[1] - 2.5)**2
```

```python
cons = ({'type': 'ineq', 'fun': lambda x:  x[0] - 2 * x[1] + 2},
        {'type': 'ineq', 'fun': lambda x: -x[0] - 2 * x[1] + 6},
        {'type': 'ineq', 'fun': lambda x: -x[0] + 2 * x[1] + 2})
```

```python
bnds = ((0, None), (0, None))
```

```python
res = minimize(fun, (2, 0), method='SLSQP', bounds=bnds,
               constraints=cons)
```
