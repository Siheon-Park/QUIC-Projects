```python
from pseudo_swap_classifier import pseudo_empirical_SWAP_classifier, pseudo_SWAP_classifier
from classifier import SVM
from datageneration import iris_data
import numpy as np

X, y, Xt, yt = iris_data(2**6)
Xin = np.array([X[2*i] for i in range(int(len(y)/2))])
Xout = np.array([X[2*i+1] for i in range(int(len(y)/2))])
yin = np.array([y[2*i] for i in range(int(len(y)/2))])
yout = np.array([y[2*i+1] for i in range(int(len(y)/2))])

pesvm = pseudo_empirical_SWAP_classifier(Xin, yin, 1, 5)
psvm = pseudo_SWAP_classifier(X, y, 1, 5)
svm = SVM(X, y, 100)

svm.optimize(np.random.rand(svm.num_data))
psvm.optimize(np.random.rand(len(psvm.theta)))
pesvm.optimize(np.random.rand(len(pesvm.theta)), Xout, yout)
```

```python
import matplotlib.pyplot as plt
asdf='sv'
a=(0, 1)

plt.figure()
svm.plot(option=asdf, a=a)
plt.figure()
psvm.plot(option=asdf, a=a)
plt.figure()
pesvm.plot(option=asdf, a=a)
```

```python
print(svm.check_performance(Xt, yt), psvm.check_performance(Xt, yt), pesvm.check_performance(Xt, yt))
```
