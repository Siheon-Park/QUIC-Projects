# Balanced dataset

```python
from classifiers.datasets.dataloader import Example_4x2
from classifiers.visualization import Plot_Data
from classifiers.convex.svm import BinarySVM
from classifiers.kernel import Kernel
import numpy as np
from matplotlib import pyplot as plt
```

```python
plt.rc('font', size=10) # controls default text sizes 
plt.rc('axes', titlesize=15) # fontsize of the axes title 
plt.rc('axes', labelsize=15) # fontsize of the x and y labels 
plt.rc('xtick', labelsize=15) # fontsize of the tick labels 
plt.rc('ytick', labelsize=15) # fontsize of the tick labels 
plt.rc('legend', fontsize=15) # legend fontsize 
plt.rc('figure', titlesize=20) # fontsize of the figure title
```

```python
X, y = Example_4x2(balanced=True)()
plt.scatter(X[y==0,0], X[y==0,1], c='r', label='y=0')
plt.scatter(X[y==1,0], X[y==1,1], c='b', label='y=1')
plt.grid()
plt.legend(loc='upper left')
plt.xlabel('$x_0=θ$')
plt.ylabel('$x_1=φ$')
plt.title('Single Qubit Bloch Sphere')
```

```python
svm = BinarySVM(kernel=Kernel('SingleQubit'), mutation='REDUCED_QASVM', C=None, k=0.1)
svm.fit(X, y)
plt.plot(svm.f(X), 'k')
plt.axhline(0, ls='--', c='k')
plt.xticks(range(4), [f'Data {i}' for i in range(4)])
plt.ylabel('$f(x)$')
plt.yticks(np.linspace(-0.5, 0.5, 5))
plt.title('Reduced QASVM ($k=0.1$)')
plt.grid()
```
