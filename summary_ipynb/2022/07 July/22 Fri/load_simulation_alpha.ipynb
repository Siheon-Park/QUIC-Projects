```python
import sys
sys.path.extend(['/home/quic/QUIC-Projects/'])
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from pathlib import Path
import json
```

```python
dir_name = 'numerical_simulation_NN_2'
n_qubits = 13
n_layers = 19
```

```python
def load_alpha(dir_name, n_qubits, n_layers):

    dir_path = Path('/home/quic/QUIC-Projects/summary_ipynb/2022/07 July/22 Fri') / dir_name
    sub_dir_path = dir_path / f'n_qubits={n_qubits}' / f'n_layers={n_layers}'
    with open(sub_dir_path/'result.json') as f:
        result = json.load(f)
    with open(sub_dir_path/'reference.json') as f:
        reference = json.load(f)
    true_alpha = np.array(reference['alpha'])
    est_alpha = np.array(result['alpha'])

    return true_alpha, est_alpha
```

```python
true_alpha, est_alpha = load_alpha(dir_name, n_qubits, n_layers)
```

```python
x = np.arange(len(true_alpha))
plt.bar(x, true_alpha, width=0.3, label='Theory')
plt.bar(x+0.5, est_alpha, width=0.3, label='Estimated')
plt.axhline(1/len(true_alpha), color='black', linestyle='dashed', label='uniform')
plt.legend()
```
