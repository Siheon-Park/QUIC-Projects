# Reduced vs normal along k

```python
np.random.seed(10)
dl = Toy2DLinearLoader(w=(1, -1), b=100)
X, y = dl(num_data=100, noise=0)
#X, y = make_moons(100, noise=0.1)
ds = DataScaler('standard', with_std=False)
Plot_Data(X, y).plot()
```

```python
ks = [0.01, 100.0]
Cs = [0.1, 10.0]
```

```python
fig, axes = plt.subplots(2, 4, figsize=(24, 8))
ax = axes[0, 0]; C=Cs[0]; k=ks[0]
svm = BinarySVM(Kernel('linear'), C=C, k=k, mutation='REDUCED_SVM')
svm.fit(X, y)
Plot_SVM(svm).plot_boundary(ax=ax, fig=fig, color_setting={-1:'b', 0:'k', 1:'r'})
ax.set_title(f'Reduced SVM, C={C}, k={k}')

ax = axes[0, 1]; C=Cs[0]; k=ks[1]
svm = BinarySVM(Kernel('linear'), C=C, k=k, mutation='REDUCED_SVM')
svm.fit(X, y)
Plot_SVM(svm).plot_boundary(ax=ax, fig=fig, color_setting={-1:'b', 0:'k', 1:'r'})
ax.set_title(f'Reduced SVM, C={C}, k={k}')

ax = axes[1, 0]; C=Cs[1]; k=ks[0]
svm = BinarySVM(Kernel('linear'), C=C, k=k, mutation='REDUCED_SVM')
svm.fit(X, y)
Plot_SVM(svm).plot_boundary(ax=ax, fig=fig, color_setting={-1:'b', 0:'k', 1:'r'})
ax.set_title(f'Reduced SVM, C={C}, k={k}')

ax = axes[1, 1]; C=Cs[1]; k=ks[1]
svm = BinarySVM(Kernel('linear'), C=C, k=k, mutation='REDUCED_SVM')
svm.fit(X, y)
Plot_SVM(svm).plot_boundary(ax=ax, fig=fig, color_setting={-1:'b', 0:'k', 1:'r'})
ax.set_title(f'Reduced SVM, C={C}, k={k}')


ax = axes[0, 2]; C=Cs[0]; k=ks[0]
svm = BinarySVM(Kernel('linear'), C=C, k=k, mutation='REDUCED_primal_SVM')
svm.fit(X, y)
Plot_SVM(svm).plot_boundary(ax=ax, fig=fig, color_setting={-1:'b', 0:'k', 1:'r'})
ax.set_title(f'Reduced (primal) SVM, C={C}, k={k}')

ax = axes[0, 3]; C=Cs[0]; k=ks[1]
svm = BinarySVM(Kernel('linear'), C=C, k=k, mutation='REDUCED_primal_SVM')
svm.fit(X, y)
Plot_SVM(svm).plot_boundary(ax=ax, fig=fig, color_setting={-1:'b', 0:'k', 1:'r'})
ax.set_title(f'Reduced (primal) SVM, C={C}, k={k}')

ax = axes[1, 2]; C=Cs[1]; k=ks[0]
svm = BinarySVM(Kernel('linear'), C=C, k=k, mutation='REDUCED_primal_SVM')
svm.fit(X, y)
Plot_SVM(svm).plot_boundary(ax=ax, fig=fig, color_setting={-1:'b', 0:'k', 1:'r'})
ax.set_title(f'Reduced (primal) SVM, C={C}, k={k}')

ax = axes[1, 3]; C=Cs[1]; k=ks[1]
svm = BinarySVM(Kernel('linear'), C=C, k=k, mutation='REDUCED_primal_SVM')
svm.fit(X, y)
Plot_SVM(svm).plot_boundary(ax=ax, fig=fig, color_setting={-1:'b', 0:'k', 1:'r'})
ax.set_title(f'Reduced (primal) SVM, C={C}, k={k}')

plt.savefig('./figs/reduced_svm_vs_reduced_primal_svm.png')
```
