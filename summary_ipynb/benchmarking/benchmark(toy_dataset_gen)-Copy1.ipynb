```python
import numpy as np 
from matplotlib import pyplot as plt
```

```python
plt.rcParams["font.size"] = 12
plt.rcParams["figure.figsize"] = (5, 4)
```

```python
from classifiers.convex.svm import BinarySVM
from classifiers.kernel import Kernel
kernel = Kernel('SingleQubit')
svm = BinarySVM(kernel, C=None, k=10, mutation='REDUCED_QASVM')
```

```python
from classifiers.datasets.dataloader import Example_4x2
X, y = Example_4x2(True)()
```

```python
plt.scatter(X[y==0,0], X[y==0, 1], c='r', label='y=0')
plt.scatter(X[y==1,0], X[y==1,1], c='b', label='y=1')
plt.legend()
plt.grid()
plt.xlabel('$x_0=θ$')
plt.ylabel('$x_1=φ$')
plt.title('Single Qubit Bloch Sphere')
plt.savefig('./figs/ToyDataDistrubution_Balanced.svg')
```

```python
svm.fit(X, y)
plt.plot(svm.f(X), 'o-k', label='regression')
plt.hlines(0, xmin=0, xmax=3, linestyle='--', colors='k')
plt.grid()
plt.xticks(range(4), [f'Data {n}' for n in range(4)])
plt.yticks([-.5, -.25, 0, .25, .5])
plt.ylabel('$f(X)$')
plt.title('Reduced QASVM ($k=10$)')
plt.legend()
plt.savefig('./figs/ToyDatasetRegression_Balanced.svg')
```
